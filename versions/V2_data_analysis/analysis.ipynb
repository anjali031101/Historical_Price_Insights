```python
cleaned_data = home_data.dropna()
```

```python
cleaned_data
```

```python
cleaned_data['1 Month Change'] = cleaned_data['1 Month Change'].str.replace('\xa0', '').str.rstrip('%').astype(float)
cleaned_data['12 Month Change'] = cleaned_data['12 Month Change'].str.replace('\xa0', '').str.rstrip('%').astype(float)
cleaned_data['Year to Date Change'] = cleaned_data['Year to Date Change'].str.replace('\xa0', '').str.rstrip('%').astype(float)
```

```python
cleaned_data['1 Month Change'] = cleaned_data['1 Month Change'].replace('', np.nan)
cleaned_data['12 Month Change'] = cleaned_data['12 Month Change'].replace('', np.nan)
cleaned_data['Year to Date Change'] = cleaned_data['Year to Date Change'].replace('', np.nan)
```

```python
home_data['1 Month Change'] = home_data['1 Month Change'].str.replace('\xa0', '')
```

```python
condition = home_data['1 Month Change'] == '\xa0'
```

```python
df_filtered = home_data[~condition]
```

```python
df_filtered
```

```python
cleaned_file_path = 'cleaned_commodities.csv'
df_filtered.to_csv(cleaned_file_path, index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization code
sns.set(style="whitegrid")

# Plot Monthly Averages
plt.figure(figsize=(14, 7))
sns.barplot(x='Monthly Avg', y='Commodity Name', data=df_filtered.sort_values(by='Monthly Avg', ascending=False))
plt.title('Monthly Average of Commodity Prices')
plt.xlabel('Monthly Average Price')
plt.ylabel('Commodity Name')
plt.tight_layout()
```

```python
# Plot 1 Month Change
plt.figure(figsize=(14, 7))
sns.barplot(x='1 Month Change', y='Commodity Name', data=df_filtered.sort_values(by='1 Month Change', ascending=False))
plt.title('1 Month Change in Commodity Prices')
plt.xlabel('1 Month Change (%)')
plt.ylabel('Commodity Name')
plt.tight_layout()
```

```python
# Plot 12 Month Change
plt.figure(figsize=(14, 7))
sns.barplot(x='12 Month Change', y='Commodity Name', data=df_filtered.sort_values(by='12 Month Change', ascending=False))
plt.title('12 Month Change in Commodity Prices')
plt.xlabel('12 Month Change (%)')
plt.ylabel('Commodity Name')
plt.tight_layout()
```
